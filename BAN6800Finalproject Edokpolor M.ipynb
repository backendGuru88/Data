```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv(r"C:\Users\Ogbemudia\Downloads\consumer_electronics_sales_data.csv")

# Display the column names to check for Customer_Age
print("Columns in DataFrame:", df.columns)

# Create Age Group feature if Customer_Age exists
if 'Customer_Age' in df.columns:
    def categorize_age(age):
        if age < 20:
            return 'Teen'
        elif 20 <= age < 30:
            return 'Young Adult'
        elif 30 <= age < 50:
            return 'Adult'
        else:
            return 'Senior'

    df['Age_Group'] = df['Customer_Age'].apply(categorize_age)
else:
    print("Customer_Age column not found!")

# Create Discounted Price feature
if 'Product_Price' in df.columns and 'Discount' in df.columns:
    df['Discounted_Price'] = df['Product_Price'] * (1 - df['Discount'] / 100)

# Log transformation for Product Price
if 'Product_Price' in df.columns:
    df['Log_Product_Price'] = np.log(df['Product_Price'] + 1)

# Scale the Customer Age feature
if 'Customer_Age' in df.columns:
    scaler = StandardScaler()
    df['Scaled_Age'] = scaler.fit_transform(df[['Customer_Age']])

# Create Customer Lifetime Value feature
if 'Total_Spent' in df.columns and 'Total_Orders' in df.columns:
    df['Customer_Lifetime_Value'] = df['Total_Spent'] / (df['Total_Orders'] + 1)

# Check for NaN values
print(df.isnull().sum())

# Drop any columns that are no longer needed
# Update the column names here based on your DataFrame
df = df.drop(columns=['Customer_Age', 'Total_Spent', 'Discount'], errors='ignore')

# Display the updated DataFrame
df.head()
```
